```python
import matplotlib.pyplot as plt
import csv
```

```python
year = []
WO = []
consumption = []

with open('istherecorrelation.csv', 'r') as csvfile:
    data = csv.reader(csvfile, delimiter=";") 
    
    #skip the header 
    next(data)
    
    # collect data values    
    for row in data:
        year.append(row[0])
        WO.append(row[1])
        consumption.append(row[2])
```

```python
# plot the results

plt.plot(year, WO, label = "WO students in the Netherlands (x 1000)")
plt.plot(year, consumption, label = "Beer consumption (x 1000 HL)")
plt.xlabel("Year")
plt.legend()
plt.savefig("Correlation_proof", dpi = 300)
plt.show()
```
